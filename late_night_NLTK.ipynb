```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.sentiment import SentimentIntensityAnalyzer
import string
import pandas as pd
import string
from nltk import word_tokenize
from nltk.sentiment import SentimentIntensityAnalyzer

# Load the dataframe
df = pd.read_csv('https://raw.githubusercontent.com/SamPasco2024/LateNightCharisma/main/Late-Night%20Interview%20Charisma%20-%20Sheet1.csv')

# Define a function to remove punctuation and lowercase the text
def clean_text(text):
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Lowercase the text
    text = text.lower()
    return text

# Apply the function to the transcript column and create a new column "edited transcript"
df['edited transcript'] = df['Transcript'].apply(clean_text)

# Define a function to get the most frequent word
def most_frequent_word(text):
    # Tokenize the text
    tokens = word_tokenize(text)
    # Get the frequency distribution of the tokens
    freq_dist = nltk.FreqDist(tokens)
    # Return the most common word
    return freq_dist.most_common(1)[0][0]

# Apply the function to the edited transcript column and create a new column "Most frequent word"
df['Most frequent word'] = df['edited transcript'].apply(most_frequent_word)

# Define a function to get the number of "um", "um's", and "uh's" per 100 words
def uhs_per_100_words(text):
    # Tokenize the text
    tokens = word_tokenize(text)
    # Get the frequency distribution of the tokens
    freq_dist = nltk.FreqDist(tokens)
    # Get the number of "um", "um's", and "uh's"
    uhs = freq_dist['um'] + freq_dist["um's"] + freq_dist['uh']
    # Calculate the number of uhs per 100 words
    uhs_per_100_words = (uhs / len(tokens)) * 100
    return uhs_per_100_words

# Apply the function to the edited transcript column and create a new column "Uh's per 100 words"
df['Uh\'s per 100 words'] = df['edited transcript'].apply(uhs_per_100_words)

# Define a function to get the sentiment of the text
def get_sentiment(text):
    # Initialize the sentiment analyzer
    sid = SentimentIntensityAnalyzer()
    # Get the sentiment scores
    sentiment_scores = sid.polarity_scores(text)
    # Return the overall sentiment
    return sentiment_scores['compound']

# Apply the function to the edited transcript column and create a new column "Overall sentiment"
df['Overall sentiment'] = df['edited transcript'].apply(get_sentiment)
```
